### Crop Pictures and Saving

```python
import cv2
import matplotlib.pyplot as plt
import os

pictures_folder = 'E:\Guan-Ming\Deep_ML/test_data/'
new_pictures_folder = 'E:\Guan-Ming\Deep_ML/test_data_resize/'
```

```python
files = os.listdir(pictures_folder)
pictures = []
for f in files:
    if 'IMG' in f:
        pictures.append(f)
```

```python
pictures[144]
```

```python
index = 155
#pic = cv2.imread(pictures_folder+pictures[index])
pic = cv2.imread(pictures_folder+'3_IMG_5161.jpg')
pic_plt = pic[:,:,::-1]
#pic_plt = cv2.resize(pic_plt, (400,256), interpolation=cv2.INTER_AREA)
pic_plt = cv2.resize(pic_plt, (256,400), interpolation=cv2.INTER_AREA)
plt.imshow(pic_plt)
plt.axis('off')
plt.show()
pic_plt.shape
```

```python
shift_h = 80
shift_w = 15
size = 215
pic_plt_crop = pic_plt[shift_h:shift_h+size, shift_w:shift_w+size,:]
pic_plt_crop.shape
```

```python
plt.imshow(pic_plt_crop)
plt.axis('off')
plt.show()
```

```python
pic = cv2.resize(pic_plt_crop[:,:,::-1], (256,256), interpolation=cv2.INTER_AREA)
#cv2.imwrite(new_pictures_folder+'resized_'+pictures[index], pic)
#cv2.imwrite(new_pictures_folder+'resized_'+pictures[index].replace('png', 'jpg'), pic)
cv2.imwrite(new_pictures_folder+'resized_'+'3_IMG_5161.jpg', pic)
```
